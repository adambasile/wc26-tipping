```python
from selenium import webdriver
from selenium.webdriver.common.by import By

import re
from dataclasses import dataclass
from enum import Enum
```

```python
class MatchResult(Enum):
    WIN = 1
    DRAW = 0
    LOSS = -1

@dataclass
class GroupStageTip:
    name: str
    player_id: str
    home: str
    away: str
    h_score: str
    a_score: str

    @property
    def result(self):
        h_i = int(self.h_score)
        a_i = int(self.a_score)

        if h_i == a_i:
            return MatchResult.DRAW
        elif h_i > a_i:
            return MatchResult.WIN
        else:
            return MatchResult.LOSS

    @property
    def match_identifier(self):
        return f"{self.home.replace(" ", "_")}-{self.away.replace(" ", "_")}"
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://wc2026-tipping-nu.vercel.app/leaderboard")
```

```python
rows = driver.find_elements(by=By.CLASS_NAME, value="lb-row")
```

```python
row = rows[0]
```

```python
name = row.find_element(by=By.CLASS_NAME, value="lb-name").text
button = row.find_element(by=By.TAG_NAME, value="button")
on_click = button.get_attribute("onClick")
player_id = re.match(R"viewPlayerTips\('(.+)'\)", on_click).group(1)
```

```python
button.click()
```

```python
view_modal = driver.find_element(by=By.CLASS_NAME, value="view-modal-box")
```

```python
close_button = view_modal.find_element(by=By.CLASS_NAME, value="view-modal-close")
```

```python
content_group = view_modal.find_element(by=By.ID, value="vmcontent-group")
```

```python
gs_tips = [GroupStageTip(name=name, player_id=player_id, home=home, away=away, h_score=h_score, a_score=a_score) for home, h_score, a_score, away in re.findall(R"(.+)\n(\d+) – (\d+)\n(.+)", content_group.text)]
```
