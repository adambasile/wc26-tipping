```python
import pandas as pd
```

```python
raw_data = pd.read_csv("https://www.eloratings.net/World.tsv", sep="\t", header=None)
teams_lookup = pd.read_csv("00_eloratings_teams.csv")
```

```python
out = pd.DataFrame(index=raw_data.index)
out["team"] = raw_data[2].map(teams_lookup.set_index("code")["team"])
out["rating"] = raw_data[3]
```

```python
out.to_csv("01_ratings.csv", index=False)
```
